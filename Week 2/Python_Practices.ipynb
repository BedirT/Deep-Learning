```python
import numpy as np
from scipy import stats
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = [13, 15, 16, 16, 19, 20, 20, 21, 22, 22, 
        25, 25, 25, 25, 30,33, 33, 35, 35, 35, 
        35, 36, 40, 45, 46, 52, 70]
```

```python
print("mean: " + str(np.mean(data)))
print("median: " + str(np.median(data)))
print("mode: " + str(stats.mode(data)[0]))
```

```python
print("midrange: " + str((max(data) + min(data)) / 2))
```

```python
print("First Quartile: " + str(np.percentile(data, 25)))
print("Third Quartile: " + str(np.percentile(data, 75)))
```

```python
s = pd.Series(data)
s.describe()
```
